```python
def distance(x1,x2):
    suma=x1-x2
    suma=suma**2
    return (sum(suma))**0.5
```

```python
def generate_points(dims=2, n=2):
    import numpy as np
    lista=[np.random.rand(dims,) for i in range(n)]
    return lista
```

```python
def run_experiment(lista_dims, n_experiments):
    import matplotlib.pyplot as plt
    import seaborn as sb
    import numpy as np
    n = len(lista_dims)
    fig,axs=plt.subplots(1,n,figsize=(20,5))    
    for j,dims in enumerate(lista_dims):
        distancias = []
        for i in range(n_experiments):
            points=generate_points(dims)
            distancias.append(distance(points[0],points[1]))
        distancias = np.array(distancias)
        
        #Calculamos algunas medidas de centralidad y dispersion y Graficamos un histograma.
        mu = distancias.mean()
        median = np.median(distancias)
        sigma = distancias.std()
        textstr = '\n'.join((
            r'$\mu=%.2f$' % (mu, ),
            r'$\mathrm{mediana}=%.2f$' % (median, ),
            r'$\sigma=%.2f$' % (sigma, )))

        axs[j].hist(distancias, 30, color='crimson')
        title = 'No. de dimensiones: '+str(dims)
        axs[j].set_title(title, fontdict=dict(weight='bold', color='darkred', size=14))
        props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
        axs[j].text(0.05, 0.95, textstr, transform=axs[j].transAxes, fontsize=14,
                verticalalignment='top', bbox=props)
        
```

```python
run_experiment([2,5,10,10000],1000)
```

### Es interesante que a mayor dimensionalidad, mayor es la distancia promedio entre 2 puntos elegidos al azar. Esto tambien significa que es mas probable que un modelo sobreajuste los datos de entrenamiento, puesto que una instancia nueva (un ejemplo nuevo) probablemente sera lejano a los datos usados para entrenar el modelo.
