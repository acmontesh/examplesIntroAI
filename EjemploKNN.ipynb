```python
data_dict = {'Nombre':['Cobra','Cascabel','Boa Constrictora','Pollo','Rana Venenosa','Cebra','Caiman','Piton','Pez Millon'],
            'Pone_huevos':[1,1,0,1,1,0,1,1,0],
            'Escamas':[1,1,1,0,0,0,1,1,1],
            'Venenoso':[1,1,0,0,1,0,0,0,0],
            'Sangre_fria':[1,1,1,0,0,0,1,1,0],
            'Numero_patas':[0,0,0,2,4,4,4,0,0],
            'Reptil':[1,1,1,0,0,0,1,1,0]}
df=pd.DataFrame.from_dict(data_dict)
df
```

```python
def calculate_distance_matrix(X):
    import pandas as pd
    import numpy as np
    cols=X.columns
    id_vector=X[cols[0]]
    vectors=[]
    for i in range(len(id_vector)):
        v = X.loc[i][1:-1].values
        vectors.append(v)
    dm = np.zeros((len(vectors),len(vectors)))
    for i in range(len(vectors)):
        for j in range(len(vectors)):
            dm[i,j]=dist(vectors[i],vectors[j])
    return dm

def dist(x1,x2):
    suma=0
    for i in range(len(x1)):
        suma=suma+(x1[i]-x2[i])**2
    suma=suma**0.5
    return suma
```

```python
df2=pd.DataFrame(calculate_distance_matrix(df), index=df['Nombre'])
df2.columns=df['Nombre']
df2
```
